```python
import matplotlib.pyplot as plt
import pandas as pd

# create a dataframe with the given data
data = {'Education Level': ['No certificate, diploma or degree', 'High school diploma or equivalent', 'Postsecondary certificate, diploma or degree', 'Apprenticeship or trades certificate or diploma', 'College, CEGEP or other non-university certificate or diploma', 'University certificate or diploma below bachelor level', 'University certificate, diploma or degree at bachelor level or above', "Bachelor's degree", 'University certificate, diploma or degree above bachelor level'],
        'Number of Individuals': [9755, 9505, 12555, 2335, 4980, 700, 4540, 3665, 315],
        'Percentage': [30.70, 29.90, 39.50, 7.30, 15.70, 2.20, 14.30, 11.50, 1.00]}
df = pd.DataFrame(data)

# create a color map for the different education levels
colors = plt.get_cmap('Set1')

# plot a horizontal bar chart with each category represented by a different color and bar label
fig, ax = plt.subplots(figsize=(10, 8))
for i, (edu_level, percentage) in enumerate(zip(df['Education Level'], df['Percentage'])):
    ax.barh(i, percentage, color=colors(i), label=edu_level)
    ax.text(percentage, i, f"{percentage:.2f}%", ha='left', va='center')
ax.set_yticks(range(len(df)))
ax.set_yticklabels(df['Education Level'])
ax.set_xlabel('Percentage')
ax.set_title('Educational Attainment')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
income_ranges = ['$0-$5k', '$5k-$10k', '$10k-$15k', '$15k-$20k', '$20k-$25k', '$25k-$30k', '$30k-$35k', '$35k-$40k', '$40k-$45k', '$45k-$50k', '$50k-$60k', '$60k-$70k', '$70k-$80k', '$80k-$90k', '$90k-$100k', '$100k-$125k', '$125k-$150k', '$150k-$200k', '$200k+']
counts = [370, 515, 770, 950, 990, 865, 870, 885, 815, 655, 1305, 1285, 1070, 900, 810, 1090, 545, 390, 95]

# Create a gradient color scheme
colors = plt.cm.get_cmap('YlGnBu')(np.linspace(0, 1, len(income_ranges)))

# Create the horizontal bar chart
plt.barh(income_ranges, counts, color=colors)

# Add labels and title
plt.xlabel('Number of People')
plt.ylabel('Income Ranges')
plt.title('Distribution of Incomes')

# Add bar labels
for i, v in enumerate(counts):
    plt.text(v + 20, i, str(v), color='black')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data
labels = ['Under $5,000', '$5,000 to $9,999', '$10,000 to $14,999', '$15,000 to $19,999', '$20,000 to $24,999', 
          '$25,000 to $29,999', '$30,000 to $34,999', '$35,000 to $39,999', '$40,000 to $44,999', '$45,000 to $49,999', 
          '$50,000 to $59,999', '$60,000 to $69,999', '$70,000 to $79,999', '$80,000 to $89,999', '$90,000 to $99,999', 
          '$100,000 to $124,999', '$125,000 to $149,999', '$150,000 to $199,999', '$200,000 and over']
values = [370, 515, 770, 950, 990, 865, 870, 885, 815, 655, 1305, 1285, 1070, 900, 810, 1090, 545, 390, 95]

# Define the colormap
colors = plt.cm.Reds(np.linspace(0.4, 1, len(labels)))

# Create the bar chart
fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(labels, values, color=colors)

# Set the chart title and axis labels
ax.set_title('Salary Distribution', fontsize=16)
ax.set_xlabel('Salary Range', fontsize=14)
ax.set_ylabel('Number of Respondents', fontsize=14)

# Set the x-axis tick rotation
plt.xticks(rotation=90)

# Add the bar labels
for i, v in enumerate(values):
    ax.text(i, v+20, str(v), ha='center', fontsize=12)

# Show the chart
plt.show()
```
